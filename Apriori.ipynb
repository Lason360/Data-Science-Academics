```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
```

```python
columns_with_1_unique_value = supermarket_df.columns[supermarket_df.nunique()==1]
```

```python
columns_with_1_unique_value
```

```python
df = supermarket_df.drop(columns=columns_with_1_unique_value)
```

```python
df_augmented = df.replace({'t': 1, '?': 0})
```

```python
df_high = df_augmented.loc[df_augmented['total'] == 'high']
df_low = df_augmented.loc[df_augmented['total'] == 'low']
```

```python
df_high = df_high.drop(columns=['total'])
df_low = df_low.drop(columns=['total'])
```

```python
frequent_itemsets = apriori(df_high, min_support=0.8, use_colnames=True)

# Convert support to frequency
frequent_itemsets['frequency'] = frequent_itemsets['support'] * len(df)

print("Frequent Itemsets with their Frequency:")
print(frequent_itemsets[['itemsets', 'frequency']])
```

```python
frequent_itemsets
```

```python
# Apply the Apriori algorithm
frequent_itemsets = apriori(df_high, min_support=0.6, use_colnames=True)

# Filter itemsets with exactly 3 items
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
two_itemsets = frequent_itemsets[frequent_itemsets['length'] == 2]

# Convert support to frequency
two_itemsets['frequency'] = two_itemsets['support'] * len(df_high)

print("Frequent 3-itemsets with their Frequency:")
print(two_itemsets[['itemsets', 'frequency']])
```

```python
df_2_apriori = two_itemsets.sort_values(by='support')
```

```python
# Apply the FP-Growth algorithm
frequent_itemsets = fpgrowth(df_high, min_support=0.5, use_colnames=True)

# Filter itemsets with exactly 3 items
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
three_itemsets = frequent_itemsets[frequent_itemsets['length'] == 3]

# Convert support to frequency
three_itemsets['frequency'] = three_itemsets['support'] * len(df_high)

# Sort the itemsets by frequency
three_itemsets_sorted = three_itemsets.sort_values(by='frequency', ascending=False)

print("Frequent 3-itemsets sorted by frequency:")
print(three_itemsets_sorted[['itemsets', 'frequency']])
```
